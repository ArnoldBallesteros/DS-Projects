```python
df = original.replace('?',np.NaN)
df = df.replace(' ',np.NaN)
```

```python
for i in df.columns:
    print(str(i) + " " + "{0:.2f}".format(df[str(i)].isnull().sum()/df.shape[0]*100) + "% Missing")
```

```python
#7/16/18 - Next steps, continue dropping NaN columns.
```

```python
df = df.drop(columns=['weight','payer_code','medical_specialty'],axis=1)
```

```python
age_range = df.age.unique()
age_nml = [5,15,25,35,45,55,65,75,85,95]
```

```python
for i in range(len(age_range)):
    df['age'] = df['age'].replace(age_range[i],age_nml[i])
```

```python
df['diag_1'] = pd.notnull(df['diag_1'])
df['diag_2'].fillna(' ',inplace = True)
df['diag_3'].fillna(' ',inplace = True)
```

```python
df[pd.notnull(df['race'])]['race'].unique()
```

```python
df['race'] = df['race'].replace('Asian',0)
df['race'] = df['race'].replace('Caucasian',1)
df['race'] = df['race'].replace('AfricanAmerican',2)
df['race'] = df['race'].replace('Hispanic',3)
df['race'] = df['race'].replace('Other',4)
```

```python
df['race'] = df['race'].fillna(int(round(df['race'].mean())))
```
